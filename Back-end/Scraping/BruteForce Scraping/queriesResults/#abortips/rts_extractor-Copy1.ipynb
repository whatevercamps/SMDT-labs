```python
def getRTS(tws):
    #obtener retweets por id
    rwts = []
    print("procesando rts para ", len(tws), "tweets")
    for tweet in tws:
        rts = []
        try: 
            if "tweet_id" in tweet:
                rts = [t._json for t in api.retweets(tweet["tweet_id"], tweet_mode="extended")]
                print("for tweet id", tweet["tweet_id"])
            elif "id" in tweet:
                rts = [t._json for t in api.retweets(tweet["id"], tweet_mode="extended")]
                print("for tweet id", tweet["id"])
            else:
                print("no tiene el atributo id, pasando al siguiente")
                continue
            for d in rts:
                rwts.append({
                    "id": d["id"], 
                    "user_followers": d["user"]["followers_count"], 
                    "retweets": 0, 
                    "likes": 0, 
                    "user_id": d["user"]["id"],
                    "username": "@{}".format(d["user"]["screen_name"]),
                    "user_location": d["user"]["location"],
                    "user_followers_count": d["user"]["followers_count"],
                    "user_created_at": d["user"]["created_at"],
                    "user_favourites_count": d["user"]["favourites_count"],
                    "user_statuses_count": d["user"]["statuses_count"],
                    "user_verified": d["user"]["verified"],
                    "url": "https://twitter.com/{}/status/{}".format(d["user"]["screen_name"], d["id"]),
                    "texto": d["full_text"],
                    "fecha": corregir(d["created_at"])[0],
                    "timestamp": corregir(d["created_at"])[1],
                    "es_retweet": 1,
                    "es_quote": 0,
                    "retweeted_id": d["retweeted_status"]["id"],
                    "quoted_tweet": "NA"
                })
                if "retweeted_status" in d:
                    cambiar = [t for t in tws if str(t["id"]) == str(d["retweeted_status"]["id_str"])]
                    for c in cambiar:
                        c["user_followers_count"] = d["retweeted_status"]["user"]["followers_count"]
                        c["user_location"] = d["retweeted_status"]["user"]["location"]
                        c["user_created_at"] = d["retweeted_status"]["user"]["created_at"]
                        c["user_favourites_count"] = d["retweeted_status"]["user"]["favourites_count"]
                        c["user_statuses_count"] = d["retweeted_status"]["user"]["statuses_count"]
                        c["user_verified"] = d["retweeted_status"]["user"]["verified"]

            print("rts obtenidos", len(rts))
        except tweepy.RateLimitError as e:
            print("límite de Twitter alcanzado, esperando 60 secs".format(e))
            time.sleep(60)
            continue
        except tweepy.TweepError as e:
            print("Tweepy error:{}".format(e))
            break
        except StopIteration:
            break
        except Exception as e:
            print("Error desconocido: {}".format(e), sys.exc_info())
            break
    return rwts
```

```python
def getQuotes(tws):
    print("procesando qts para ", len(tws), "tweets")
    #menthod for getting quoutes for tweets and retweets
    quotes = []
    for dat in tws:
        tweets = []
        Tweets_df = None
        c = twint.Config()
        c.Search = dat["url"]
        c.User_full = True
        c.Profile_full = True
        c.Retweets = True
        c.Store_object_tweets_list = tweets
        # Run
        twint.run.Search(c)
        for tweet in tweets:
            quotes.append({
                "id": tweet.id_str, 
                "retweets": tweet.retweets_count,
                "likes": tweet.likes_count,
                "user_id": tweet.user_id_str,
                "username": "@{}".format(tweet.username),
                "user_location": "NA",
                "user_followers_count": "NA",
                "user_created_at": "NA",
                "user_favourites_count": "NA",
                "user_statuses_count": "NA",
                "user_verified": "NA",
                "url": tweet.link,
                "texto": tweet.tweet,
                "fecha": tweet.datastamp,
                "timestamp": tweet.datetime,
                "es_retweet": 0,
                "es_quote": 1,
                "retweeted_id": "NA",
                "quoted_tweet": dat["url"].split("/")[-1]
            })
        print("qts for tweet url ", dat["url"])
        print("qts obtenidos", len(tweets))
        print("")
    return quotes
```

```python
#recursive method for boths 
def getTodo(tweets):
    
    if(tweets is None or len(tweets) == 0):
        return []
    else:
        rts = getRTS(tweets)
        quotes = getQuotes(tweets)

        todo = rts + quotes
        return todo + getTodo(quotes)
```
