```python
def getUrl(node):
    try:
        return "https://twitter.com/{0}/status/{1}".format(node["data"]["user"]["screen_name"],
                                                               node["id"])
        
    except:
        try:
            return node["data"]["entities"]["media"][0]["url"]
        except:
            return node["data"]["extended_entities"]["media"][0]["url"] 
```

```python
from datetime import datetime
def corregir(x):
    try:
        tm = int(x)
        return time.strftime('%d/%m/%Y'.format(x%1000), time.gmtime(x/1000.0))
    except:
        format_string = "%d/%b/%Y"
        dsplit = x.split(" ")
        date_object = datetime.strptime("{}/{}/{}".format(dsplit[2], dsplit[1], dsplit[-1]), format_string)
        return date_object.strftime("%d/%m/%Y")
```
