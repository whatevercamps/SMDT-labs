```python
file_in = open("todos-noticia-2.json", "r")
rwts = json.load(file_in)
```

```python
nodes = {}
for rt in rwts:
    tw = rt["retweeted_status"]
    tw["es_retweet"] = 0
    rt["es_retweet"] = 1
    rt["retweeted_id"] = tw["id"]
    nodes[tw["id"]] = tw
    nodes[rt["id"]] = rt
```

```python
nodes = [{"id": n, "data": nodes[n]} for n in nodes]
```
